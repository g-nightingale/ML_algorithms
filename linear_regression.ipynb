**Purpose:** Build a simple linear regression model using gradient descent in numpy

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline

n = 100
epochs = 10
alpha = 0.05
```

```python
# Create x and y variables
x = np.random.randn(n, 1)
y = x + np.random.randn(n, 1)/2.0
```

```python
# Run optimisation
# Initalise weight and bias terms
w = 0.001
b = 0.001

plt.figure(figsize=(12,10))

for epoch in range(epochs):
    
    if epoch==0:
        y_hat = (x * w) + b
        plt.plot(x, y_hat, label='untrained')


    # Initialise partial derivatives to 0 for each epoch
    dl_dw = 0.0
    dl_db = 0.0
    
    # Update partial derivaties for each example
    for i in range(x.shape[0]):
        dl_dw += 2 * x[i] * ((x[i] * w + b) - y[i])
        dl_db += 2 * ((x[i] * w + b) - y[i])
        
    w = w - alpha * dl_dw * 1/x.shape[0]
    b = b - alpha * dl_db * 1/x.shape[0]
    
    # Create updated prediction and plot it
    y_hat = (x * w) + b
    
    plt.plot(x, y_hat, label='epoch: ' + str(epoch))
    
plt.scatter(x, y)
plt.legend(loc='upper left')
plt.show()
```
